```python
import pandas as pd
pd.set_option('max_columns', None)
```

```python
df = pd.read_csv("Amazon EC2 Instance Comparison.csv")
df.rename(columns = {'API Name': 'Instance Type'}, inplace = True)
df.head()
```

```python
df1 = pd.read_csv("ec2.txt" , sep="\t|\s{2,}", index_col = False, header=0, engine='python')
df1.head()
```

```python
#merging
df3 = pd.merge(df, df1, on='Instance Type', how='inner')
```

```python
#extracting useful columns from merged dataframe
df3 = df3[['Name', 'Instance Type', 'Memory_x', 'Compute Units (ECU)', 'vCPUs_x', 'GiB of Memory per vCPU', 'GPUs', 
           'GPU memory', 'ECU per vCPU', 'Price', 'Monthly', 'Spot Price']]
#renaming
df3.rename(columns={'Memory_x':'Memory',  'vCPUs_x':'vCPUs'}, inplace=True)

df3.set_index('Instance Type', inplace=True)
```

```python
df3.head()
```

```python
#adding utilization columns
df3['Avg_CPU'] = df3['Max_CPU'] = df3['Avg_Mem'] = df3['Max_Mem'] = df3['Avg_GPU'] = df3['Max_GPU']  = ''
```

```python
#reading utilization for instance = t2.medium
t2_med = pd.read_csv('utils/util-t2med.txt', sep='\s', names=['CPU', 'Memory'], usecols=(1,2), engine='python')

df3.at['t2.medium', 'Avg_CPU'] = t2_med['CPU'].mean()
df3.at['t2.medium', 'Max_CPU'] = t2_med['CPU'].max()
df3.at['t2.medium', 'Avg_Mem'] = t2_med['Memory'].mean()
df3.at['t2.medium', 'Max_Mem'] = t2_med['Memory'].max()
```

```python
#reading utilization for instance = t2.large
t2_lar = pd.read_csv('utils/util-t2large.txt', sep='\s', names=['CPU', 'Memory'], usecols=(1,2), engine='python')

df3.at['t2.large', 'Avg_CPU'] = t2_lar['CPU'].mean()
df3.at['t2.large', 'Max_CPU'] = t2_lar['CPU'].max()
df3.at['t2.large', 'Avg_Mem'] = t2_lar['Memory'].mean()
df3.at['t2.large', 'Max_Mem'] = t2_lar['Memory'].max()
```

```python
#reading utilization for instance = t2.xlarge
t2_xlar = pd.read_csv('utils/util-t2xlarge.txt', sep='\s', names=['CPU', 'Memory'], usecols=(1,2), engine='python')

df3.at['t2.xlarge', 'Avg_CPU'] = t2_xlar['CPU'].mean()
df3.at['t2.xlarge', 'Max_CPU'] = t2_xlar['CPU'].max()
df3.at['t2.xlarge', 'Avg_Mem'] = t2_xlar['Memory'].mean()
df3.at['t2.xlarge', 'Max_Mem'] = t2_xlar['Memory'].max()
```
